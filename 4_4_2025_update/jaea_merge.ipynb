```python
# IMPORT ALL THE NECESSARY LIBRARIES
#
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh
import seaborn as sns
from pathlib import Path
from sklearn.metrics import r2_score
import sys
import os
#
# IDENTIFY THE DEFAULT PATH FOR ALL OUTPUT FILES (change the variable for path_1 to whatever your filepath is):
#
path_1 = Path(r'C:\Users\zavarin1\Documents\codes\v9_25_2024')
# Pull in lscie dataset, csv
#
df_lscie = pd.read_csv(path_1.joinpath("sc.dataset.csv"),low_memory=False)
df_lscie_refs = pd.read_csv(path_1.joinpath("Reference.csv"),low_memory=False)
res = pd.merge(df_lscie, df_lscie_refs, how='left', on=['Reference'])
df_lscie = pd.DataFrame(res)
#drop useless columns
df_lscie.drop('comments', inplace=True, axis=1)
df_lscie.drop('Imported by', inplace=True, axis=1)
df_lscie.drop('ID', inplace=True, axis=1)
df_lscie.drop('X_axis', inplace=True, axis=1)
df_lscie.drop('X_val', inplace=True, axis=1)
df_lscie.drop('X_SD', inplace=True, axis=1)
df_lscie.drop('X_units', inplace=True, axis=1)
df_lscie.drop('Y_axis', inplace=True, axis=1)
df_lscie.drop('Y_val', inplace=True, axis=1)
df_lscie.drop('Y_SD', inplace=True, axis=1)
df_lscie.drop('Y_units', inplace=True, axis=1)
#df_lscie.drop('Temp_SD', inplace=True, axis=1)
#pd.set_option('display.width', 200)
print(df_lscie.head(2))
```

```python
# Pull in jaea dataset, comma delimited
#
df_jaea = pd.read_csv(path_1.joinpath("jaea.csv"),low_memory=False)
print(df_jaea.head(2))
```
